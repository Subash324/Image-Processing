```python
image=cv2.imread("/content/drive/MyDrive/5th sem/image processing/photos/Copy of messi.png")
cv2_imshow(image)
```

```python
image_gray=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
cv2_imshow(image_gray)
```

```python
image_bw= cv2.bitwise_not(image_gray)
cv2_imshow(image_bw)
```

```python
(thresh, blackAndWhiteImage) = cv2.threshold(image_gray,100, 150, cv2.THRESH_BINARY)
cv2_imshow(blackAndWhiteImage)
```

```python
image_edge=cv2.Canny(image_gray,100,250)
cv2_imshow(image_edge)
```

```python
image1=cv2.imread("/content/drive/MyDrive/5th sem/image processing/photos/Copy of book (1).jpg")
```

```python
grayscaled = cv2.cvtColor(image1,cv2.COLOR_BGR2GRAY)
retval, threshold = cv2.threshold(grayscaled, 10, 255, cv2.THRESH_BINARY)
cv2_imshow(image1)
cv2_imshow(threshold)
```

```python
threshold = cv2.adaptiveThreshold(grayscaled, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 115, 1)
cv2_imshow(image1)
cv2_imshow(threshold)
```

```python
image2=cv2.imread("/content/drive/MyDrive/5th sem/image processing/photos/Copy of contrast.png")
cv2_imshow(image2)
```

```python
original = image2.copy()
```

```python
xp = [0, 64, 128, 192, 255]
fp = [0, 16, 128, 240, 255]
```

```python
x = np.arange(256)
```

```python
table = np.interp(x, xp, fp).astype('uint8')
```

```python
image2 = cv2.LUT(image2, table)
```

```python
cv2_imshow( original)
cv2_imshow(image2)
```
