```python
import cv2 as cv
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
img= cv.imread("/content/drive/MyDrive/5th sem/image processing/photos/erosion.png",0)
```

```python
#---------------------------------Erosion Operation-----------------------
```

```python
kernel = np.ones((5,5),np.uint8)
erosion = cv.erode(img,kernel,iterations = 2)
cv2_imshow(img)
cv2_imshow(erosion)
```

```python
#------------------------------Dialation Operation-----------------------------
```

```python
img1= cv.imread("/content/drive/MyDrive/5th sem/image processing/photos/dialation.png",0)
dilation = cv.dilate(img1,kernel,iterations = 1)
cv2_imshow(img1)
cv2_imshow(dilation)
```

```python
#----------------------opening Operation-------------------
```

```python
close= cv.imread("/content/drive/MyDrive/5th sem/image processing/photos/open.png",0)
opening = cv.morphologyEx(close, cv.MORPH_OPEN, kernel)
cv2_imshow(close)
cv2_imshow(opening)
```

```python
#----------------------Closing Operation-------------------------
```

```python
close= cv.imread("/content/drive/MyDrive/5th sem/image processing/photos/close.png",0)
closing = cv.morphologyEx(close, cv.MORPH_CLOSE, kernel)
cv2_imshow(close)
cv2_imshow(closing)
```

```python
#--------------------------morphological Gradient---------------
```

```python
fingerprint= cv.imread("/content/drive/MyDrive/5th sem/image processing/photos/img.jpg",0)
gradient = cv.morphologyEx(fingerprint, cv.MORPH_GRADIENT, kernel)
cv2_imshow(fingerprint)
cv2_imshow(gradient)
```
