```python
import cv2 as cv
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
img=cv.imread("/content/drive/MyDrive/5th sem/image processing/photos/close.png",0)
```

```python
cv2_imshow(img)
```

```python
# eroison
```

```python
kernel = np.ones((5,5),np.uint8)
erosion = cv.erode(img,kernel,iterations = 1)
cv2_imshow(img)
cv2_imshow(erosion)
```

```python
dilation = cv.dilate(erosion,kernel,iterations = 3)
cv2_imshow(erosion)
cv2_imshow(dilation)
```

```python
close= cv.imread("/content/drive/MyDrive/5th sem/image processing/photos/open.png",0)
opening = cv.morphologyEx(close, cv.MORPH_OPEN, kernel)
cv2_imshow(close)
cv2_imshow(opening)
```
